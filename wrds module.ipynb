# WRDS 数据查询

## Method 1. Using get_table( )

Parameters:
+ **library**
+ **table**
+ **columns** (optional)
+ **obs**: the number of rows (optional)
+ **offset**: the starting point (optional)

```python
data= db.get_table(library = 'djones', table = 'djdaily', columns=['date','dji'],obs=10)
data
```

## Method 2. Using raw_sql( )  (most useful way)

Parameters:
+ **sql**: the SQL string
+ **date_cols**: column to parse as date (optional)
+ **index_col**: columns to set as index (optional)

```python
data=db.raw_sql('SELECT date, dji FROM djones.djdaily LIMIT 10;', date_cols=['date'])
data
```

## Method 3. Using get_row_count( )  （计算Dataset行数）

Parameters:
+ **library** 
+ **table**

```python
data = db.get_row_count('djones', 'djdaily')
data
```

# Joining Data from Separate Datasets 从多个数据库同时提取

```python
db.raw_sql('''SELECT a.gvkey, a.datadate, a.tic, a.conm, a.at, a.lt, b.prccm, b.cshoq 
              FROM comp.funda a JOIN comp.secm b 
              ON a.gvkey = b.gvkey AND a.datadate = b.datadate
              WHERE a.tic='IBM' AND a.datafmt='STD' AND a.indfmt='INDL' ''')
```

# Passing Parameters to SQL

```python
parm = {'tickers':('0015B', '0030B', '0032A', '0033A', '0038A')}
data = db.raw_sql('''SELECT datadate, gvkey, cusip 
                     FROM comp.funda
                     WHERE tic in %(tickers)s''', params=parm)
data
```
